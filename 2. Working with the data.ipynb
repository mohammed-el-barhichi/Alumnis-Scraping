```python
import pandas as pd
```

```python
#xlsx
data = pd.read_excel('Alumni Scraping.xlsx', sheet_name='te')
```

```python
data.head()
```

```python
# copy data
df = data.copy()
```

```python

# drop last two columns

df.drop(['Column5', 'Column6'], axis=1, inplace=True)
```

```python
# convert comanies column dattype to list

df['Companies'] = df['Companies'].apply(lambda x: x.split(',,'))
df['Roles'] = df['Roles'].apply(lambda x: x.split(',,'))

```

```python
df
```

```python
# convert float objects in the locations column to text

df['Locations'] = df['Locations'].astype(str)
```

```python

df['Locations'] = df['Locations'].apply(lambda x: x.split(',,'))
```

```python
df.dtypes
```

```python
df
```

```python
df['Locations'][2]
```

```python
# leave only the last element not equal to ' ' in the 3 columns

df['Locations'] = df['Locations'].apply(lambda x: x[-1] if x[-1] != ' ' else x[-2])
df['Roles'] = df['Roles'].apply(lambda x: x[-1] if x[-1] != ' ' else x[-2])
df['Companies'] = df['Companies'].apply(lambda x: x[-1] if x[-1] != ' ' else x[-2])
```

```python
df
```

```python
# save as excel

df.to_excel('One Company.xlsx', sheet_name='te', index=False)
```
